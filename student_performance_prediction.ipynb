***Heatmap***

**Problem:** A university wants to analyze correlations between studentsâ€™ study hours, attendance %, and exam scores.

**Columns:** study_hours, attendance, exam_score

**Steps:**

Collect data from students (CSV or DataFrame).

Calculate correlation matrix (.corr()).

Use sns.heatmap() to visualize correlations.

Check which variable has the strongest link with exam scores.

```python
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv('/content/student_performance_dataset.csv')
```

```python
data=pd.DataFrame(data)
data
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt

numeric_data = data.select_dtypes(include=['number'])
sns.heatmap(numeric_data.corr(), annot=True, fmt='.2f', cbar=True, cmap="coolwarm", square=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
correlation_with_exam_score = numeric_data.corr()['Final_Exam_Score'].sort_values(ascending=False)
print("Correlation with Final_Exam_Score:")
print(correlation_with_exam_score)
```
